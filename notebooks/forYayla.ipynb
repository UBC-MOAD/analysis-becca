# Collection of SalishSeaCast Data for Yayla

```python
import xarray as xr
from pathlib import Path
import numpy as np
import datetime as dt
from functools import partial
```

```python
path = Path('/results2/SalishSea/nowcast-green.202111')
day = [dt.datetime(2014,1,1)+dt.timedelta(days=i) for i in range(int(365*10+2))] # 2014-2023 (inclusive)
subdir = ["{:%d%b%y}".format(d).lower() for d in day]
```

```python
# filepaths
chemT = [sorted(path.glob("{}/SalishSea_1d_{:%Y%m%d}_{:%Y%m%d}_chem_T.nc".format(subdir[i], 
                                                                                day[i], 
                                                                                day[i]))) 
                                                                                for i in range(len(day))]
```

```python
drop_vars = ("PAR", "turbidity")

def _preprocess(file, x_bnds, y_bnds, d_bnds):
    return file.sel(deptht=slice(*d_bnds), y=slice(*y_bnds), x=slice(*x_bnds))
y_bnds, x_bnds, d_bnds = (761.5,762,5), (137.5,138.5), (0,1)
partial_func = partial(_preprocess, x_bnds=x_bnds, y_bnds=y_bnds, d_bnds=d_bnds)

data = xr.open_mfdataset(chemT,drop_variables=drop_vars,preprocess=partial_func)
```
